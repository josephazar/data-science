## Compressed Sensing demo
###  sampling 25% of the temporal signal

```python
cr=0.25 #compression ratio
n=len(ecg)
m=int(n*cr) #we want only to keep 10% of the data
print(n,m)
```

```python
# random sample of indices
ri = np.random.choice(n, m, replace=False)
ri.sort()
ecg_comp = ecg[ri]
```

```python
plt.plot(ecg_comp)
```

### In order to reconstruct the signal, I used the approach presented in this blog page: 
#### http://www.pyrunner.com/weblog/2016/05/26/compressed-sensing-python/

### The image below is taken from this page and explains how the Compressed Sensing works:

![Compressed sensing](../pictures/cspyrunner.png)

```python
A = spfft.idct(np.identity(n), norm='ortho', axis=0)
A = A[ri]  
vx = cvx.Variable(n)
objective = cvx.Minimize(cvx.norm(vx, 1))
constraints = [A@vx == ecg_comp]
prob = cvx.Problem(objective, constraints)
result = prob.solve(verbose=True)
x = np.array(vx.value)
x = np.squeeze(x)
ecg_rec = spfft.idct(x, norm='ortho', axis=0)
print(ecg_rec.shape)
```

```python
plt.plot(ecg_rec)
```

## ECG enhancement using UNet
### 1- Scaling the data

```python
from sklearn.preprocessing import MinMaxScaler

data=data.reshape(len(data),1)
scaler = MinMaxScaler()
data=scaler.fit_transform(data)
data=data.reshape(-1)
```

### 2- Preparing the input and output of the AutoEncoder
#### Input: Reconstructed ECG signal after being sampled with a 20% sampling rate
#### Output: Original ECG signal

```python
def get_reconstructed(comp_data,n,ri):
    A = spfft.idct(np.identity(n), norm='ortho', axis=0)
    A = A[ri]  
    vx = cvx.Variable(n)
    objective = cvx.Minimize(cvx.norm(vx, 1))
    constraints = [A@vx == comp_data]
    prob = cvx.Problem(objective, constraints)
    result = prob.solve(verbose=False)
    x = np.array(vx.value)
    x = np.squeeze(x)
    ecg_rec = spfft.idct(x, norm='ortho', axis=0)
    return ecg_rec
    
def generate_ecg_inpout(data,length=Fs):
    X_in=list()
    X_out=list()
    i=0
    step=length
    while i + length <= len(data):
        tmp=data[i:i+length]
        X_out.append(tmp)
        #compress
        cr=0.2 #compression ratio
        n=len(tmp)
        m=int(n*cr)
        ri = np.random.choice(n,m, replace=False)
        ri.sort()
        ecg_comp = tmp[ri]        
        #reconstruct
        ecg_rec=get_reconstructed(ecg_comp,n,ri)
        X_in.append(ecg_rec)
        i+=step
    X_out = np.array(X_out).reshape(len(X_out), length)
    X_in = np.array(X_in).reshape(len(X_in), length)
    return X_in,X_out
```

```python
X_in,X_out=generate_ecg_inpout(data)
```

```python
print(X_in.shape)
print(X_out.shape)
```

### 3- Creating the train and test sets

```python
X_out_test=X_out[X_out.shape[0]-5:]
X_in_test=X_in[X_in.shape[0]-5:]
X_out=X_out[:-5]
X_in=X_in[:-5]
```

### 4- Define and compile the model

```python
from tensorflow.keras.models import Model,load_model
from tensorflow.keras.layers import  Input, Concatenate, BatchNormalization,Conv1D, MaxPooling1D, UpSampling1D
import tensorflow as tf


def down(input_layer, filters, pool=True):
    conv1 = Conv1D(filters, (3, ), padding='same', activation='relu')(input_layer)
    residual = Conv1D(filters, (3, ), padding='same', activation='relu')(conv1)
    residual = BatchNormalization()(residual)
    if pool:
        max_pool = MaxPooling1D()(residual)
        return max_pool, residual
    else:
        return residual

def up(input_layer, residual, filters):
    filters=int(filters)
    upsample = UpSampling1D()(input_layer)
    upconv = Conv1D(filters, kernel_size=(2, ), padding="same")(upsample)
    upconv = BatchNormalization()(upconv)
    concat = Concatenate(axis=2)([residual, upconv])
    conv1 = Conv1D(filters, (3,), padding='same', activation='relu')(concat)
    conv2 = Conv1D(filters, (3,), padding='same', activation='relu')(conv1)
    return conv2

def get_unet_model(filters=64):
    # Make a custom U-nets implementation.
    filters = 64
    input_layer = Input(shape = [496, 1])
    #layers = [input_layer]
    residuals = []

    # Down 1, 248
    d1, res1 = down(input_layer, filters)
    residuals.append(res1)
    filters *= 2

    # Down 2, 124
    d2, res2 = down(d1, filters)
    residuals.append(res2)
    filters *= 2

    # Down 3, 62
    d3, res3 = down(d2, filters)
    residuals.append(res3)
    filters *= 2

    # Down 4, 31
    d4, res4 = down(d3, filters)
    residuals.append(res4)
    filters *= 2

    # Down 5, 31
    d5 = down(d4, filters, pool=False)

    # Up 1, 31
    up1 = up(d5, residual=residuals[-1], filters=filters/2)
    filters /= 2

    # Up 2,  62
    up2 = up(up1, residual=residuals[-2], filters=filters/2)
    filters /= 2

    # Up 3, 124
    up3 = up(up2, residual=residuals[-3], filters=filters/2)
    filters /= 2

    # Up 4, 248
    up4 = up(up3, residual=residuals[-4], filters=filters/2)
    out = Conv1D(filters=1, kernel_size=(1,), activation="sigmoid")(up4)

    model = Model(input_layer, out)
    return model
```

```python
model = get_unet_model(filters=64)
model.summary()
model.compile(optimizer='adam', loss='binary_crossentropy')
```

### 5- Train and save the model
### make sure to increase the number of epochs; I used 25 just for demonstration purpose

```python
epochs=25
history = model.fit(X_in, X_out,
                epochs=epochs,
                batch_size=2,
                shuffle=True,
                validation_data=(X_in_test, X_out_test))
model.save('UnetEcgAE_CS.h5')
```

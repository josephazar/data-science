# Sampling with Bayes classifier
This code implements a bayes classifier that generate some samples from the mnist dataset

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

We will be using multivariate_normal (A multivariate normal random variable) for drawing samples

```python
from scipy.stats import multivariate_normal as mvn
```

Now we will define the sampling class based on Bayes classifier

```python
class BayesClassifier:
    #No constructor
    def fit(self,X,Y):
        self.num_classes=len(set(Y)) #This is the number of classes
        #we initialize a list ot hold our gaussian information
        self.gaussians = []
        #The two parameters that describe a gaussian are the mean and covariance
        #For each class:
        for c in range(self.num_classes):
            #get the data belonging to this class
            #Shape of Xc: (m,28*28) where m is the number of samples
            Xc=X[Y==c]
            #find the mean
            #shape fo mean: (28*28)
            mean=np.mean(Xc,axis=0)
            #find the covariance
            #shape of cov: (28*28,28*28)
            cov=np.cov(Xc.T)
            #Append the values to a dictionary
            dic_val={'mean':mean,'cov':cov}
            self.gaussians.append(dic_val)
    
    #A function to draw a sample from a given class y
    def draw_sample_y(self,y):
        #get corresponding dictionary
        val=self.gaussians[y]
        #rvs function generates a sample (no clamping)
        #return mvn.rvs(mean=val["mean"],cov=val["cov"])
        
        #rvs function generates a sample + clamping
        return self.clamping_sample(mvn.rvs(mean=val["mean"],cov=val["cov"]))
        
    #A function that generates a random sample
    def ran_sample(self):
        y=np.random.randint(self.num_classes)
        return self.draw_sample_y(y)
    
    #In computer graphics, clamping is the process of limiting a position to an area.
    #To put clamping into perspective, pseudocode for clamping is: 
    #function: if X > Max then X := Max; if X < Min then X := Min; return X;
    #In our case, we are limiting the values in a sample to the range 0-1
    def clamping_sample(self,x):
      x = np.minimum(x, 1)
      x = np.maximum(x, 0)
      return x
```

## Load the data

```python
from keras.datasets import mnist
(X, Y), _ = mnist.load_data()
#convert from unsigned ints to floats
X = X.astype('float32')
#scale from [0,255] to [0,1]
X= X / 255.0
X=X.reshape(X.shape[0],X.shape[1]*X.shape[2])
print(X.shape)
print(Y.shape)
```

Creating an instance of our classifier and fitting the data

```python
generator=BayesClassifier()
generator.fit(X,Y)
```

Show one sample for each class and the mean image learned

```python
def plot_gen_sample(y):
    img=generator.draw_sample_y(y).reshape(28,28)
    mean=generator.gaussians[y]["mean"].reshape(28,28)
    
    plt.subplot(1,2,1)
    plt.imshow(img,cmap='gray')
    plt.title('Generated sample')
    plt.subplot(1,2,2)
    plt.imshow(mean,cmap='gray')
    plt.title('Mean image')
    plt.show()
```

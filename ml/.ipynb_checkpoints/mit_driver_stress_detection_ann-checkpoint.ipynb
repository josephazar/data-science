### Feature extraction / Extraction de caractéristiques
#### In order to extract features from the ECG signal, we are going to use the heartPy library

```python
import heartpy as hp
Fs=496 #sampling frequency

#run analysis
wd, m = hp.process(hp.scale_data(dataset.iloc[0:20000,0].values), Fs)

#visualise in plot of custom size
plt.figure(figsize=(12,4))
hp.plotter(wd, m)

#display computed measures
for measure in m.keys():
    print('%s: %f' %(measure, m[measure]))
    
```

```python
# comparing the calculated heart rate value to the mean heart rate value present in the data file

print("calculated : %f" % m['bpm'])
print("original : %f" %np.mean(dataset.iloc[0:20000,3].values))
```

##### You can notice that the difference is acceptable

#### In order to extract features from Galvanic Skin Response signal, we are going to use cvxEDA: A Convex Optimization Approach
#### https://github.com/lciti/cvxEDA/blob/master/src/cvxEDA.py

```python
from Utils import cvxEDA
cvxEDA.process(dataset.iloc[0:20000,1].values, 1./Fs)
print()
#display computed measures
for measure in cvxEDA.eda_features.keys():
    print('%s: %f' %(measure, cvxEDA.eda_features[measure]))
```

#### Now we want to split the dataset into windows, and from each window extract the features from ECG and GSR
#### nous voulons diviser l'ensemble de données en fenêtres

```python
# We will consider a window of 40 seconds
window_size =  int(Fs*40) 
dataset=dataset[0:int((int(len(dataset)/window_size)-1)*window_size)]

print(dataset.shape)
```

```python
#dataframe containing the features to be used for classification
#the last column is the label
dataframe_columns=['bpm', 'ibi', 'sdnn',
                       'sdsd', 'rmssd', 'pnn20', 'pnn50', 'hr_mad',
                       'mean_gsr', 'mean_scl', 'slope_scl', 'mean_scr',  'max_scr', 'resp',
                       'Stress']


#in case a window contains rows having different labels, the assigned label will be of the class with max number of rows
def getCommonValue(vec):
    counts = np.bincount(vec)
    return np.argmax(counts)
```

#### This code will take some time to be executed

```python
features={}
data_list=[]

#we will use a sliding window with step = 8
    
i=0
while i <= len(dataset)-window_size:
    features={}
    window_data=dataset[i:i+window_size]  
    
    #ECG
    wd, m = hp.process(hp.scale_data(window_data["ECG"].values), Fs)
    features.update(m)

    #EDA (from hand only)
    cvxEDA.process(window_data["GSR_H"], 1./Fs)
    features.update(cvxEDA.eda_features)

    #Respiration
    features["median_resp"]=np.median(window_data["RESP"])
    features["Stress"]=getCommonValue(window_data['STRESS'])


    data_list.append([features["bpm"],
                      features["ibi"],
                      features["sdnn"],
                      features["sdsd"],
                      features["rmssd"],
                      features["pnn20"],
                      features["pnn50"],
                      features["hr_mad"],
                      features["mean_gsr"],
                      features["mean_scl"],
                      features["slope_scl"],
                      features["mean_scr"], 
                      features["max_scr"],            
                      features["median_resp"],
                      features["Stress"]

            ])
    i=i+int(window_size/8)

    
```

```python
driverdf = pd.DataFrame(data_list,columns = dataframe_columns) 
```

### Classification task starts here

#### Check if null values exist

```python
import seaborn as sns
sns.heatmap(driverdf.isnull(),yticklabels=False,cbar=False,cmap='Blues')
```

##### We can notice that the dataframe contains null values, we will drop them

```python
driverdf = driverdf.dropna(axis = 0, how ='any') 
sns.heatmap(driverdf.isnull(),yticklabels=False,cbar=False,cmap='Blues')
```

```python
X= driverdf.iloc[:,0:14].values
y= driverdf.iloc[:,14].values

print(X.shape)
print(y.shape)
```

```python
print(pd.Series(y).value_counts())
```

```python
#shuffle the data
from sklearn.utils import shuffle
X,y=shuffle(X, y,  random_state=7) 
```

```python
from sklearn.model_selection import train_test_split
#Splitting the dataset into the Training set and Test set
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)


print(pd.Series(y_train).value_counts())
print(pd.Series(y_test).value_counts())
```

#### Features scaling

```python
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
X_train=sc_X.fit_transform(X_train)
X_test=sc_X.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential  # Initialize our NN 
from tensorflow.keras.layers import Dense # Build the different layers in our NN
from tensorflow.keras.layers import Dropout


# define baseline model
def baseline_model():
    classifier = Sequential()
    classifier.add(Dense(activation='relu',input_dim=14,units=9,kernel_initializer='uniform'))
    classifier.add(Dropout(rate=0.1))
    classifier.add(Dense(activation='relu',units=9,kernel_initializer='uniform'))
    classifier.add(Dropout(rate=0.1))
    classifier.add(Dense(activation='relu',units=9,kernel_initializer='uniform'))
    classifier.add(Dropout(rate=0.1))
    classifier.add(Dense(activation='relu',units=9,kernel_initializer='uniform'))
    classifier.add(Dropout(rate=0.1))
    classifier.add(Dense(activation='sigmoid',units=1,kernel_initializer='uniform'))
    classifier.compile(optimizer= 'adam', loss = 'binary_crossentropy',metrics=['acc'])
    return classifier



classifier = baseline_model()
```

```python
classifier.fit(X_train,y_train,batch_size=10,epochs=100)
```

```python
# Test on unseen data
results = classifier.evaluate(X_test, y_test)
print('Final test set loss: {:4f}'.format(results[0]))
print('Final test set accuracy: {:4f}'.format(results[1]))
```

#### Cross validation

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold


estimator = KerasClassifier(build_fn=baseline_model, epochs=100, batch_size=10, verbose=0)
kfold = KFold(n_splits=10, shuffle=True, random_state=7)
results = cross_val_score(estimator, X_train, y_train, cv=kfold)
print("Baseline: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```
